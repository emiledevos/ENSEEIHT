# Etude des paramètres $\lambda_k$ et $\mu_k$

Dans un premier temps, nous allons étudier comment les paramètres lambda et mu varie en fonction des itérations efectuer par l'algorithme du lagrangien augmenté. Dans un premier temps, traçons cette courbe pour la fonction 2 est le point x03

```julia
include("../test/fonctions_de_tests.jl")
include("../src/Lagrangien_Augmente.jl")
using Plots

tol = 1e-4
x_sol, f_sol, flag, nb_iters, m, l = Lagrangien_Augmente(fct2, grad_fct2, hess_fct2, contrainte2, grad_contrainte2, hess_contrainte2, x03, tol_abs=tol, tol_rel=tol, algo_noc="newton")

e= range(1, length(m), length=length(m))
p1 = plot(xlabel= "n", ylabel = "mu", e,m)
e= range(1, length(l), length=length(l))
p2 = plot(xlabel= "n", ylabel = "lambda", e,l)
plot(p1,p2)
```

Nous allons maintenant tracé ce même graphique pour la fonction $f_1$ avec le point inital x01.

```julia
include("../test/fonctions_de_tests.jl")
include("../src/Lagrangien_Augmente.jl")
using Plots

tol = 1e-4
x_sol, f_sol, flag, nb_iters, m, l = Lagrangien_Augmente(fct1, grad_fct1, hess_fct1, contrainte1, grad_contrainte1, hess_contrainte1, x01, tol_abs=tol, tol_rel=tol, algo_noc="rc-cauchy")


e= range(1, length(m), length=length(m))
p1 = plot(xlabel= "n", ylabel = "mu", e,m)

e= range(1, length(l), length=length(l))
p2 = plot(xlabel= "n", ylabel = "lambda", e,l)
plot(p1,p2)
```

Dans un premier temps, on peut voir sur les graphiques que soit lambda varie et mu est constant, soit lambda est constant et c'est mu qui varie. Ceci s'explique par le fait que on est soit dans le cas de la pénalisation si les contraintes ne sont pas respecté, alors mu évolue mais lambda reste constant. Soit on est dans le cas ou les contraintes sont respecté, et dans ce cas on peut modifier lambda en gardant mu constant.

Dans le premier cas, nous pouvons observer que la variable mu augmente, tandis que celle de lambda diminue. Dans le second cas, on voit que mu augmente également mais que lambda augmente aussi. ceci est cohérent car mu ne peut que augmenter ou être constant. Pour lambda, ça variations dépend de $c(x)$ c'est pour cela que ce paramètre peut diminuer ou augmenter.

# Etude des variations de $\tau$

Observons maintenant l'influence de $\tau$ sur l'algorithme du lagrangien augmenté.

```julia
include("../test/fonctions_de_tests.jl")
include("../src/Lagrangien_Augmente.jl")
using Plots

tol_erreur= 1e-2
nom="cauchy"
nb_test = 10

e= range(1, 10, length=nb_test)
n=zeros(nb_test)
for i in 1:nb_test
    x_sol, f_sol, flag, n[i], m, l = 
    Lagrangien_Augmente(fct2, grad_fct2, hess_fct2, contrainte2, grad_contrainte2, hess_contrainte2, x03, tol_abs=tol,τ=e[i] ,tol_rel=tol, algo_noc="rc-cauchy")
end
plot(e,n, xlabel= "tau", ylabel="nbIter", label =false )
```

On voit sur ce plot que plus $\tau$ augmente, plus le nombre d'itération diminue. au début. Ensuite il reste constant. ON peut ainsi comprendre que plus $\tau$ est grand, plus les variations de $\lambda_k$ et $\mu_k$ sont importante. Ils semblent donc logique que la résolutions soit plus grande avec une valeur de $\tau$ plus grande. ON peut donc conclure que l'impacte de $\tau$ est positif sur l'algorithme si il est plus grand.
